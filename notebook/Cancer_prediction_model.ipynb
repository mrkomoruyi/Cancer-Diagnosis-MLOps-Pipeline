```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv(filepath_or_buffer="C:/dev/Datasets/The_Cancer_data_1500_V2.csv")

# Nominalize the dataset 
train_df = dataset.iloc[:1200, :]
test_df = dataset.iloc[1200:, :]

X_train_df = train_df.iloc[:, :8]
X_test_df = test_df.iloc[:, :8]
y_train_df = train_df.iloc[:, 8]
y_test_df = test_df.iloc[:, 8]

X_train_df_mean = X_train_df.mean()
X_train_df_std = X_train_df.std()

X_train_df_norm = (X_train_df - X_train_df_mean) / X_train_df_std
X_test_df_norm = (X_test_df - X_train_df_mean) / X_train_df_std
```

```python
X_df_norm = pd.concat(objs=(X_train_df_norm, X_test_df_norm))

y_df = pd.concat(objs=(y_train_df, y_test_df))
```

len(dataloader) gives you the number of batches in the dataloader

```python
# using sklearn to use random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score,log_loss, confusion_matrix

X = np.float64(X_df_norm.values)
y = np.float64(y_df.values)

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, shuffle=True, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
auroc = roc_auc_score(y_test, y_pred)
f1score = f1_score(y_test, y_pred)
logloss = log_loss(y_test, y_pred)
confusion_mtx = confusion_matrix(y_test, y_pred)

print(f"Accuracy = {accuracy}\n AUROC = {auroc:.3f}\n F1Score = {f1score}\n Log Loss = {logloss}\n Confusion matrix = \n{confusion_mtx}")
```
